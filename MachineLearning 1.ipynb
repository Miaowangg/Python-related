```python
def sub_divide(x):
  print('initial value',x)
  x -= 2 # x = x - 2
  return x/2, x/4

for values in df['hba1c'].iloc[::10]:
  result = sub_divide(values)
  print('result',result)
  print('')
print('after loop')
```

```python
fig = plt.figure(figsize = (10,7))
ax = fig.subplots()
ax.scatter(df['rbg'], df['hba1c'], label = "data", color = 'pink', marker = 's')

ax.set_xlabel('Random Blood Glucose (rbg) [mmol/L]', fontsize = 20)
ax.set_ylabel('Glycosylated Hemoglobin (hba1c) [DCCT %]', fontsize = 20)
ax.set_title('hba1c & rbg readings from patients at \n the Karl Bremer District Hospital', fontsize = 30)

ax.legend(bbox_to_anchor = (1,1), loc='upper right', fontsize = 25)
plt.show()

```
